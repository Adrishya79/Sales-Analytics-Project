```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "Product": ["Lenga", "Choli", "Gunyo","Cholo", "Blouse", "Saree", "EthnicWear", "Dresses", "OnePiece", "Kurta Surwal", "Daura", "Surwal"],
    "Category": ["Ethnic Wear", "Ethnic Wear", "Traditional Wear", "Traditional Wear", "Western Wear", "Ethnic Wear", "Ethnic Wear", "Western Wear", "Western Wear", "Traditional Wear", "Traditional Wear", "Traditional Wear"],
    "Units Sold": [50, 40, 30, 35, 60, 45, 55, 65, 70, 80, 25, 40],
    "Revenue": [25000, 20000, 15000, 17500, 30000, 22500, 27500, 32500, 35000, 40000, 12500, 20000],
    "Cost": [15000, 12000, 9000, 10500, 18000, 13500, 16500, 19500, 21000, 24000, 7500, 12000],
    "Profit": [10000, 8000, 6000, 7000, 12000, 9000, 11000, 13000, 14000, 16000, 5000, 8000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Plot Revenue vs. Units Sold
plt.figure(figsize=(10, 6))
plt.bar(df['Product'], df['Revenue'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation=45)
plt.show()
```
